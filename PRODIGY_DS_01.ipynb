```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset from the World Bank URL
url = 'C:\\Users\\user\\Downloads\\task1\\API_SP.POP.TOTL_DS2_en_csv_v2_305614.csv'
df = pd.read_csv(url, skiprows=4)

# Display the first few rows of the dataset to understand its structure
print(df.head())

# Select the relevant columns (Country Name and Population in the most recent year, e.g., 2022)
df_recent = df[['Country Name', '2022']].dropna()

# Rename columns for easier access
df_recent.columns = ['Country', 'Population']

# Sort the data by population in descending order
df_recent_sorted = df_recent.sort_values(by='Population', ascending=False)

# Display the sorted data
print(df_recent_sorted.head())

# Create a bar chart to visualize the population distribution of the top 10 most populous countries
plt.figure(figsize=(10, 6))
plt.bar(df_recent_sorted['Country'][:10], df_recent_sorted['Population'][:10], color='skyblue')
plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Top 10 Most Populous Countries in 2022')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Create a histogram to visualize the population distribution
plt.figure(figsize=(10, 6))
plt.hist(df_recent_sorted['Population'], bins=30, color='lightgreen', edgecolor='black')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.title('Histogram of Country Populations in 2022')
plt.tight_layout()
plt.show()

# Create a pie chart to visualize the population distribution of the top 10 most populous countries
plt.figure(figsize=(10, 6))
plt.pie(df_recent_sorted['Population'][:10], labels=df_recent_sorted['Country'][:10], autopct='%1.1f%%', colors=plt.cm.Paired.colors)
plt.title('Top 10 Most Populous Countries in 2022')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.show()
```
